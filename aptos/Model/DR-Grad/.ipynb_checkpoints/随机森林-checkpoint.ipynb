```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier   
from sklearn.ensemble import RandomForestClassifier
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve, auc 
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import sklearn.metrics as sm
%matplotlib inline
plt.rcParams['font.size'] = 24
```

```python
data = pd.read_excel("P101.xlsx")
data
```

```python
data = pd.read_csv('output_1.csv')
X= data.drop(['label'], axis=1)   # 取出用于建模的特征列X
Y = data['label']   # 取出标签列Y
```

```python
data.replace(to_replace=r'^\s*$', value=np.nan, regex=True, inplace=True)   # 把各类缺失类型统一改为NaN的形式
data.isnull().mean()
```

```python
x = data.drop(['state'], axis=1)   # 取出用于建模的特征列X
label = data['state']   # 取出标签列Y
```

```python
# 划分训练集、测试集
xtrain, xtest, ytrain, ytest = train_test_split(x, label, test_size=0.3)
xtrain.head()
```

```python
# 单颗决策树
clf = DecisionTreeClassifier(class_weight='balanced',random_state=37)   
clf = clf.fit(xtrain, ytrain)   # 拟合训练集
score_c = clf.score(xtest, ytest)   # 输出测试集准确率
 
# 随机森林
rfc = RandomForestClassifier(n_estimators=90,max_depth=1,class_weight='balanced')   
rfc = rfc.fit(xtrain, ytrain)
score_r = rfc.score(xtest, ytest)
```

```python
# 决策树 预测测试集
y_test_proba_clf = clf.predict_proba(xtest)
false_positive_rate_clf, recall_clf, thresholds_clf = roc_curve(ytest, y_test_proba_clf[:, 1])  
# 决策树 AUC指标
roc_auc_clf = auc(false_positive_rate_clf, recall_clf)     

    
# 随机森林 预测测试集
      
y_test_proba_rfc = rfc.predict_proba(xtest)
y_pred = rfc.predict(xtest)
false_positive_rate_rfc, recall_rfc, thresholds_rfc = roc_curve(ytest, y_test_proba_rfc[:, 1])  

cm = confusion_matrix(y_true=ytest, y_pred=y_pred)
# 打印混淆矩阵
print("Confusion Matrix: ")
print(cm)

cp = sm.classification_report(ytest, y_pred)
print("---------------分类报告\n", cp)
# 随机森林 AUC指标
roc_auc_rfc = auc(false_positive_rate_rfc, recall_rfc)     
 
# 画图 画出俩模型的ROC曲线
plt.plot(false_positive_rate_clf, recall_clf, color='blue', label='AUC_clf=%0.3f' % roc_auc_clf) 
plt.plot(false_positive_rate_rfc, recall_rfc, color='orange', label='AUC_rfc=%0.3f' % roc_auc_rfc)  
plt.legend(loc='best', fontsize=15, frameon=False)  
plt.plot([0, 1], [0, 1], 'r--')  
plt.xlim([0.0, 1.0])  
plt.ylim([0.0, 1.0])  
plt.ylabel('Recall')  
plt.xlabel('Fall-out')  
plt.show()
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score
print('Accuracy score:', accuracy_score(ytest, y_pred))
print('Accuracy score:', accuracy_score(ytest, y_pred))
print('Recall:', recall_score(ytest, y_pred, average='weighted'))
print('F1-score:', f1_score(ytest, y_pred, average='weighted'))
print('Precision score:', precision_score(ytest, y_pred, average='weighted'))
```

```python
#调参第一步：n_estimators
cross = []
for i  in range(0,100,10):
    rf = RandomForestClassifier(n_estimators=i+1, n_jobs=-1,random_state=42)
    cross_score = cross_val_score(rf, xtest, ytest, cv=5).mean()
    cross.append(cross_score)
plt.plot(range(1,101,10),cross)
plt.xlabel('n_estimators')
plt.ylabel('acc')
plt.show()
print((cross.index(max(cross))*10)+1,max(cross))
```

```python
#调参第一步：maxdepth
cross = []
for i  in range(1,20,1):
    rf = RandomForestClassifier(n_estimators=90,max_depth=i+1, n_jobs=-1,random_state=42)
    cross_score = cross_val_score(rf, xtest, ytest, cv=5).mean()
    cross.append(cross_score)
plt.plot(range(1,20,1),cross)
plt.xlabel('max_depth')
plt.ylabel('acc')
plt.show()
print((cross.index(max(cross))*10)+1,max(cross))
```

```python
empty = np.array([[0],[0],[0],[0],[0],[0],[0]])
n=1
while n < 51:
    y_pred = rfc.predict(xtest)   
    y_updatapred = rfc.predict(xtrain)
    from sklearn.metrics import confusion_matrix, roc_auc_score
    cm = confusion_matrix(ytest, y_pred)   
    cm_updata = confusion_matrix(ytrain, y_updatapred)
    auc = roc_auc_score(ytest, y_pred)
    auc_updata = roc_auc_score(ytrain, y_updatapred)
    a = cm[0,0]
    b = cm[0,1]
    c = cm[1,0]
    d = cm[1,1]
    sen = d/(d+c)
    sep = a/(a+b)
    a_updata = cm_updata[0,0]
    b_updata = cm_updata[0,1]
    c_updata = cm_updata[1,0]
    d_updata = cm_updata[1,1]
    sen_updata = d_updata/(d_updata + c_updata)
    sep_updata = a_updata/(a_updata + b_updata)
    first = np.array([[n],[sen],[sep],[auc],[sen_updata],[sep_updata],[auc_updata]])
    second = np.hstack((empty,first))
    empty = second
    n = n + 1
    print(n)
final_par = np.delete(second,0,axis=1)
print (final_par)
final_parT = final_par.T
np.savetxt('RFClassifier',final_parT,delimiter=',')
```
