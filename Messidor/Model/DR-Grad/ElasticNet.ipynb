```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import Lasso
from sklearn import linear_model
```

```python
data = pd.read_csv('output_7.csv')
X= data.drop('label',axis=1)
y = data['label']
```

```python
# import missingno as msno
# msno.bar(data)
```

```python
X /= X.std(axis=0)  # Standardize data (easier to set the l1_ratio parameter)
```

```python
alpharange = np.logspace(-10,0,200,base=10) 
```

```python
model_eNetcv =  ElasticNetCV(l1_ratio=0.5,alphas=alpharange,max_iter=10000,cv = 5).fit(X,y)
```

```python
model_eNetcv.alpha_
```

```python
(model_eNetcv.coef_*100).tolist()
```

```python
features_eNetcv =(model_eNetcv.coef_*100).tolist()
```

```python
data_eNet = []
for i in features_eNetcv:
    if i != 0:
        data_eNet.append(features_eNetcv.index(i))
```

```python
data_eNet
```

```python
data_eNet[0]
```

```python
plt.figure()
plt.plot(model_eNetcv.alphas_,model_eNetcv.mse_path_.mean(axis=1),marker = 'o')
plt.xlabel('Alpha')
plt.ylabel('Cross-Validation Error')
plt.title('Elastic Net Hyperparmeter Tuning')
plt.grid(True)
plt.show()
```

```python
coef_path = []
model_eNet =  ElasticNet(l1_ratio=0.7,max_iter=10000)
for alpha in alpharange:
    model_eNet.set_params(alpha=alpha)
    model_eNet.fit(X,y)
    coef_path.append(model_eNet.coef_.tolist())
```

```python
 coef_path =np.array(coef_path)
```

```python
plt.figure(dpi=300)
for i, coef in enumerate(coef_path.T):
    plt.plot(np.log10(alpharange), coef, label=f'Coefficient {i+1}')
plt.xlabel('log10(alpha)')
plt.ylabel('coefficients')
plt.title('Elastic Net Hyperparmeter Tuning')
plt.grid(True)
plt.savefig( 'Elastic Net Hyperparmeter Tuning.png')
plt.show()
```

```python
coef_path_la = []
model_la = Lasso(max_iter=10000)
for alpha in alpharange:
    model_la.set_params(alpha=alpha)
    model_la.fit(X,y)
    coef_path_la.append(model_la.coef_.tolist())
```

```python
coef_path_la=np.array(coef_path_la)
```

```python
plt.figure(dpi=300)
for i, coef in enumerate(coef_path_la.T):
    plt.plot(np.log10(alpharange), coef, label=f'Coefficient {i+1}')
plt.xlabel('log10(alpha)')
plt.ylabel('coefficients')
plt.title('lasso Hyperparmeter Tuning')
plt.grid(True)
plt.savefig( 'lasso Hyperparmeter Tuning.png')
plt.show()
```

```python
alpharange_Ri = np.logspace(-5,5,200,base=10)
```

```python
coef_path_Ri = []
model_Ri = linear_model.Ridge(max_iter=10000)
for alpha in alpharange_Ri:
    model_Ri.set_params(alpha=alpha)
    model_Ri.fit(X,y)
    coef_path_Ri.append(model_Ri.coef_.tolist())
```

```python
coef_path_Ri=np.array(coef_path_Ri)
```

```python
plt.figure(dpi=300)
for i, coef in enumerate(coef_path_Ri.T):
    plt.plot(np.log10(alpharange), coef, label=f'Coefficient {i+1}')

plt.xlabel('log10(alpha)')
plt.ylabel('coefficients')
plt.title('Ridge Hyperparmeter Tuning')
plt.grid(True)
plt.savefig( 'Ridge Hyperparmeter Tuning.png')
plt.show()
```
